## 1. Check for if there exists any null value in dataset

```python
socioeconomic_factors.isnull().sum()
```

After dropping columns that we don't need, we see that only `Childhood Blood Lead Level Screening` and `Childhood Lead Poisoning` fields are consisting 1 null value in each.

```python
socioeconomic_factors
```

Now let's look at two fields, namely 'Childhood Blood Lead Level Screening' and 'Childhood Lead Poisoning' which have consisting 1 null value in each. Since each row represents one of the community areas in Chicago city, so we can not just simply drop that two columns for the sake of our projet because blood lead level and lead poisoning may be resulted from those facilities that we are investigating in Chicago Food Inspection dataset.

Hence, we fill the null values with the median of the each fields in the dataset.

```python
socioeconomic_factors['Childhood Blood Lead Level Screening'] = \
                            socioeconomic_factors['Childhood Blood Lead Level Screening']\
                            .fillna(socioeconomic_factors['Childhood Blood Lead Level Screening'].median())
```

```python
socioeconomic_factors['Childhood Lead Poisoning'] = \
                            socioeconomic_factors['Childhood Lead Poisoning']\
                            .fillna(socioeconomic_factors['Childhood Lead Poisoning'].median())
```

Let's check now if there are still missing values in our dataset.

```python
if len(socioeconomic_factors.isnull().sum()[socioeconomic_factors.isnull().sum()!=0]) == 0:
    print("There are no null values in any columns of our socioeconomic dataset.")
```

### 2. Total Population by Chicago Community Areas

We also need the population of each community areas in Chicago for answering our research questions. Here is the [source dataset](https://www.chicagohealthatlas.org/indicators/total-population).

```python
total_population = pd.read_excel("datasets/Total_population.xlsx")
```

```python
total_population.head()
```

```python
total_population.columns
```

```python
#drop unwanted columns
total_population = total_population\
                        .drop(columns =['Category', 'SubCategory', 'Geo_ID', 'Demography', 'Demo_Group', 
                                        'Cum_Number', 'Ave_Annual_Number', 'Crude_Rate', 'Lower_95CI_Crude_Rate', 
                                        'Upper_95CI_Crude_Rate', 'Age_Adj_Rate', 'Lower_95CI_Age_Adj_Rate', 
                                        'Upper_95CI_Age_Adj_Rate', 'Percent', 'Lower_95CI_Percent', 'Indicator',
                                        'Upper_95CI_Percent', 'Weight_Number', 'Weight_Percent', 'Geography',
                                        'Lower_95CI_Weight_Percent','Upper_95CI_Weight_Percent', 'Map_Key', 'Flag'])
```

```python
total_population.dtypes
```

We could only have the dataset which is between 2012 and 2016 as the most recent dataset. We filtered our dataset based on this year range and split the Geo_Group column because we need community area and community area name as in separate columns.

```python
total_population = total_population[total_population['Year']=='2012-2016'].reset_index()
total_population = total_population.drop(columns='index')
total_population['Community Area'] = [total_population.Geo_Group[i].split('-', 1)[0] for i in range(len(total_population))] 
total_population['Community Area Name'] = [total_population.Geo_Group[i].split('-', 1)[1] for i in range(len(total_population))] 
total_population = total_population.drop(columns=['Geo_Group','Year'])
total_population = total_population[['Community Area', 'Community Area Name', 'Number']]
total_population['Community Area'] = total_population['Community Area'].astype(int)
total_population
```

```python
#Merge two dataset
full_socioeconomic_factors = socioeconomic_factors.merge(total_population, on=['Community Area', 'Community Area Name'])
full_socioeconomic_factors
```
