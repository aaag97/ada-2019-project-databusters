## 1. Check for if there exists any null value in dataset

```python
socioeconomic_factors.isnull().sum()
```

We see that only `Childhood Blood Lead Level Screening`, `Childhood Lead Poisoning`, and `Gonorrhea in Females` fields are consisting null values.

Since we do not need 'Gonorrhea' factor for our projet, we securely drop the columns related to _Gonorrhea_.

```python
#drop unwanted columns related to Gonorrhea
socioeconomic_factors = socioeconomic_factors.drop(columns =['Gonorrhea in Males', 'Gonorrhea in Females'])
```

```python
socioeconomic_factors
```

Now let's look at other two fields, namely 'Childhood Blood Lead Level Screening' and 'Childhood Lead Poisoning' which have consisting 1 null value in each. Since each row represents one of the community areas in Chicago city, so we can not just simply drop that two columns for the sake of out projet because blood lead level and lead poiaoning may be relted to those facilities that we are investigating on CFI dataset.

Hence, we fill the null values woth the median of the each fields in the dataset.

```python
socioeconomic_factors['Childhood Blood Lead Level Screening'] = \
                            socioeconomic_factors['Childhood Blood Lead Level Screening']\
                            .fillna(socioeconomic_factors['Childhood Blood Lead Level Screening'].median())
```

```python
socioeconomic_factors['Childhood Lead Poisoning'] = \
                            socioeconomic_factors['Childhood Lead Poisoning']\
                            .fillna(socioeconomic_factors['Childhood Lead Poisoning'].median())
```

Let's check now if there are still missing values in our dataset.

```python
if len(socioeconomic_factors.isnull().sum()[socioeconomic_factors.isnull().sum()!=0]) == 0:
    print("There are no null values in any columns of our socioeconomic dataset.")
```

```python
socioeconomic_factors.isna().sum().apply(lambda x: '{}% missing values'.format(100 * x/len(socioeconomic_factors)))
```
